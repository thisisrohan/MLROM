```python
class BPTT_model(Model):
    def __init__(self, rnn_model, k1, k2, k3):
        super(BPTT_model, self).__init__()
        self.rnn_model = rnn_model
        self.k1 = k1
        self.k2 = k2
        self.k3 = k3
        
    def call(self, inputs, training=None):
        return self.rnn_model.call(inputs, training)

    def train_step(self, data):

        # x, y = data
        x, y, sample_weight = data_adapter.unpack_x_y_sample_weight(data)
        
        num_timesteps = x.shape[-2]
        tbatch_len = self.k3+self.k1-1
        num_tbatches = num_timesteps // tbatch_len
        num_leftover_tsteps = num_timesteps % tbatch_len
        num_tsteps_warmup = self.k3 - self.k2

        for tbatch_idx in range(num_tbatches):
            xbegin_idx = tbatch_idx*tbatch_len
            if num_tsteps_warmup > 0:
                _ = self.rnn_model.call(x[:, xbegin_idx:xbegin_idx+num_tsteps_warmup, :], training=False)
            ytrue = y[:, xbegin_idx+self.k3-1:xbegin_idx+tbatch_len, :]
            with tf.GradientTape() as tape:
                ypred = self.rnn_model.call(x[:, xbegin_idx+num_tsteps_warmup:xbegin_idx+self.k2+self.k1-1, :], training=True)
                ypred = ypred[:, self.k2-1:, :]
                loss = self.rnn_model.compiled_loss(
                    ytrue,
                    ypred,
                    sample_weight,
                    regularization_losses=self.rnn_model.losses
                )
            self.rnn_model._validate_target_and_loss(ypred, loss)

            trainable_vars = self.rnn_model.trainable_variables
            gradients = tape.gradient(loss, trainable_vars)
            
            self.rnn_model.optimizer.apply_gradients(zip(gradients, trainable_vars))
            
            metrics = self.rnn_model.compute_metrics(x[:, xbegin_idx+num_tsteps_warmup:xbegin_idx+self.k2+self.k1-1, :], ytrue, ypred, sample_weight)

        if num_leftover_tsteps > 0:
            # just to maintain the stateful property
            _ = self.rnn_model.call(x[:, -num_leftover_tsteps:, :], training=False)

        return metrics
```

```python
def rescale_data(data, normalization_arr):
    '''
    data - [num_batches x num_timesteps x num_states]
    normalization_arr = [2 x num_states]
    '''
    new_data = data.copy()
    shape = new_data.shape
    for i in range(data.shape[-1]):
        new_data[:, i] -= normalization_arr[0, i]
        new_data[:, i] /= normalization_arr[1, i]

    return new_data

def norm_sq_time_average(data):
    data_norm_sq = np.zeros(shape=data.shape[0])
    for i in range(data.shape[1]):
        data_norm_sq[:] += data[:, i]**2
    # integrating using the trapezoidal rule
    norm_sq_time_avg = np.sum(data_norm_sq) - 0.5*(data_norm_sq[0]+data_norm_sq[-1])
    norm_sq_time_avg /= data_norm_sq.shape[0]
    return norm_sq_time_avg

def invert_normalization(data, normalization_arr):
    new_data = np.empty_like(data)
    shape = new_data.shape
    print(shape)
    for i in range(shape[-1]):
        if len(shape) == 2:
            new_data[:, i] = data[:, i]
            new_data[:, i] *= normalization_arr[1, i]
            new_data[:, i] += normalization_arr[0, i]
        elif len(shape) == 3:
            new_data[:, :, i] = data[:, :, i]
            new_data[:, :, i] *= normalization_arr[1, i]
            new_data[:, :, i] += normalization_arr[0, i]
    return new_data
```
