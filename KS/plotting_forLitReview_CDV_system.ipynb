```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from tools.misc_tools import create_CDV_data
```

```python
T = 2000
t0 = 0.0
delta_t = 0.01
params_mat = np.array([
    [0.95, -0.801*0.95, 0.1, 1.25, 0.2, 0.5]
]) # [x1*, x4*, C, beta, gamma, b]
init_state_mat = np.array([
    [0.2*(2*(i%2 - 0.5)) for i in range(6)]
])
```

```python
res_dict = create_CDV_data(
    T, t0, delta_t,
    params_mat,
    init_state_mat, return_params_arr=False,
    normalize=False,
    stddev_multiplier_for_norm=None
)
```

```python
all_data = res_dict['all_data']
```

```python
new_data = all_data[int(0//delta_t):, 0:6]
```

```python
num_cols = 1
num_rows = 6

fig = plt.figure(figsize=(7.5*num_cols, 2.5*num_rows))

input_time = np.arange(0, new_data.shape[0])*delta_t

subplot_idx = 1
for j in range(6):
    ax_orig = fig.add_subplot(num_rows, num_cols, subplot_idx)
    
#     ax_orig.plot(input_time, np.ones_like(input_time)*lorenz_attractor_pos[0, 0, j], color='C1', linewidth=0.8)
#     ax_orig.plot(input_time, np.ones_like(input_time)*lorenz_attractor_pos[0, 1, j], color='C1', linewidth=0.8)
    
#     ax_orig.axhline(lorenz_attractor_pos[0, 0, j], color='C1', linewidth=0.8)
#     ax_orig.axhline(lorenz_attractor_pos[0, 1, j], color='C1', linewidth=0.8)
    
    ax_orig.plot(input_time, new_data[:, j])#, label='Actual Data')
    ax_orig.set_ylabel(r'$x_'+str(j+1)+'$')
    ax_orig.grid(True)
    ax_orig.set_axisbelow(True)
#     if j == 0:
        # if save_figs == False or True:
#         ax_orig.title.set_text(r'$x_1^*$={:.2f},  $x_4^*$={:.2f},  $C$={:.2f}, $\beta$={:.2f}, $\gamma$={:.2f}, $b$={:.2f}'.format(
#             all_data[i*(N+1), 6],
#             all_data[i*(N+1), 7],
#             all_data[i*(N+1), 8],
#             all_data[i*(N+1), 9],
#             all_data[i*(N+1), 10],
#             all_data[i*(N+1), 11]
#         ))
#         ax_orig.legend(loc='upper left', bbox_to_anchor=[1, 1])
    subplot_idx += 1
    if j < 6-1:
        ax_orig.xaxis.set_ticklabels([])

ax_orig.set_xlabel('Time')
# plt.savefig('lorenz_time_evolution.png', dpi=300, bbox_inches='tight')
```

```python
blocked_time = np.array([
    [155, 205], 
    [418, 500],
    [1190, 1400],
])

blocked_idx = np.int32(blocked_time / delta_t)

#zonal_obj = plt.scatter(new_data[:, 0], new_data[:, 3], marker='o', s=0.5, linewidth=0.05)

begin_idx = 0
for jj in range(blocked_time.shape[0]):
    zonal_obj = plt.scatter(
        new_data[
            begin_idx:blocked_idx[jj, 0],
            0
        ],
        new_data[
            begin_idx:blocked_idx[jj, 0],
            3
        ],
        marker='o',
        s=0.5,
        linewidth=0.05,
        color='C0'
    )
    begin_idx = blocked_idx[jj, 1]
    
    blocked_obj = plt.scatter(
        new_data[
            blocked_idx[jj, 0]:blocked_idx[jj, 1],
            0
        ],
        new_data[
            blocked_idx[jj, 0]:blocked_idx[jj, 1],
            3
        ],
        marker='o',
        s=0.5,
        linewidth=0.05,
        color='C1'
    )

zonal_obj = plt.scatter(
    new_data[
        begin_idx:,
        0
    ],
    new_data[
        begin_idx:,
        3
    ],
    marker='o',
    s=0.5,
    linewidth=0.05,
    color='C0'
)

plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_4$')

plt.legend([zonal_obj, blocked_obj], ['Zonal Flow', 'Blocked Flow'], markerscale=5)

plt.grid(True)
plt.savefig('cdv_x1x4.png', dpi=300, bbox_inches='tight')
```
