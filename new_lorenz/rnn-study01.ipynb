```python
import numpy as np
import matplotlib.pyplot as plt
import os

plt.rcParams.update({
    "text.usetex":True,
    "font.family":"serif",
})
```

```python
xlabel_kwargs = {'fontsize':15}
ylabel_kwargs = {'fontsize':15}
legend_kwargs = {'fontsize':12}
title_kwargs = {'fontsize':18}
xticks_kwargs = {'fontsize':12}
yticks_kwargs = {'fontsize':12}
```

```python
cwd = os.getcwd()

esn_dname = cwd + '/saved_AR_AEESN_rnn/AR_ESN_ensemble_008'
gru_dname = cwd + '/saved_AR_AERNN_rnn/AR_rnn_007'

print('esn_dname : ', esn_dname)
print('gru_dname : ', gru_dname)
```

```python
# plot_dir = cwd + '/study-01/plots'

# if not os.path.exists(plot_dir):
#     os.makedirs(plot_dir)

esn_plot_dir = esn_dname + '/plots/study-01'
if not os.path.exists(esn_plot_dir):
    os.makedirs(esn_plot_dir)
    
gru_plot_dir = gru_dname + '/plots/study-01'
if not os.path.exists(gru_plot_dir):
    os.makedirs(gru_plot_dir)
```

```python
esn_ph_list = []
esn_numoutsteps = []
for fname in os.listdir(esn_dname):
    if fname.endswith('.npz') and fname.startswith('prediction_horizon'):
        esn_ph_list.append(esn_dname + '/' + fname)
        fname = fname[0:-4]
        fname = fname[::-1]
        idx = fname.find('--')
        fname = fname[8:idx]
        fname = fname[::-1]
        if fname == 'ZERO':
            esn_numoutsteps.append(0)
        else:
            esn_numoutsteps.append(int(fname))

print(esn_ph_list)
print(esn_numoutsteps)

esn_ph_list = np.array(esn_ph_list)
esn_numoutsteps = np.array(esn_numoutsteps)

sort_idx = np.argsort(esn_numoutsteps)
esn_numoutsteps = esn_numoutsteps[sort_idx]
esn_ph_list = esn_ph_list[sort_idx]

print('')
print(esn_ph_list)
print(esn_numoutsteps)
```

```python
gru_ph_list = []
gru_numoutsteps = []
for fname in os.listdir(gru_dname):
    if fname.endswith('.npz') and fname.startswith('prediction_horizon'):
        gru_ph_list.append(gru_dname + '/' + fname)
        fname = fname[0:-4]
        fname = fname[::-1]
        idx = fname.find('--')
        fname = fname[8:idx]
        fname = fname[::-1]
        if fname == 'ZERO':
            gru_numoutsteps.append(0)
        else:
            gru_numoutsteps.append(int(fname))

print(gru_ph_list)
print(gru_numoutsteps)

gru_ph_list = np.array(gru_ph_list)
gru_numoutsteps = np.array(gru_numoutsteps)

sort_idx = np.argsort(gru_numoutsteps)
gru_numoutsteps = gru_numoutsteps[sort_idx]
gru_ph_list = gru_ph_list[sort_idx]

print('')
print(gru_ph_list)
print(gru_numoutsteps)
```

```python
esn_ph_arr_list = []
esn_medians = []
for fname in esn_ph_list:
    with np.load(fname, allow_pickle=True) as f:
        try:
            ph_arr = f['prediction_horizon_arr']
        except:
            for key in f.keys():
                if key.find('og') != -1:
                    ph_arr = f[key]
                    break
        esn_ph_arr_list.append(ph_arr)
    num_runs = ph_arr.shape[0]
    ph_arr.sort()
    esn_medians.append(ph_arr[int(0.5*num_runs-1)])
print(esn_medians)
```

```python
def esn_plot(box_or_violin):
    yticks_min = 0
    yticks_max = int(np.max(esn_ph_arr_list)+1)
    yticks_pos = np.arange(yticks_min, yticks_max+1, 1)

    # box_or_violin = 2 # 0 == box
    #                   # 1 == violin
    #                   # 2 == box with violin super-imposed
    if box_or_violin == 0:
        parts = plt.boxplot(
            esn_ph_arr_list,
        )
    elif box_or_violin == 1:
        parts = plt.violinplot(
            esn_ph_arr_list,
            showmedians=True
        )
    elif box_or_violin == 2:
        parts = plt.violinplot(
            esn_ph_arr_list,
        )
        parts['cmaxes'].set(linewidth=0.)
        parts['cmins'].set(linewidth=0.)
        parts['cbars'].set(linewidth=0.)
        # parts['medians'].set(linewidth=0.)
        parts2 = plt.boxplot(
            esn_ph_arr_list,
        )


    plt.grid(True)
    plt.xticks(
        np.arange(esn_numoutsteps.shape[0]) + 1,
        ('Teacher\nforced',) + tuple([str(elem) for elem in esn_numoutsteps[1:]]),
        **xticks_kwargs
    )
    plt.yticks(yticks_pos, **yticks_kwargs)
    plt.xlabel('AR Training Steps', **xlabel_kwargs)
    plt.ylabel('Prediction Horizon [LT]', **ylabel_kwargs)
    plt.title('ESN Auto-regressive Training', **title_kwargs)

    if box_or_violin == 0:
        savefig_name = 'esn-study01--boxplot.pdf'
    elif box_or_violin == 1:
        savefig_name = 'esn-study01--violinplot.pdf'
    elif box_or_violin == 2:
        savefig_name = 'esn-study01--boxandviolinplot.pdf'

    plt.savefig(esn_plot_dir+'/'+savefig_name, dpi=300, bbox_inches='tight')
    plt.show()
    plt.close()
```

```python
gru_ph_arr_list = []
gru_medians = []
for fname in gru_ph_list:
    with np.load(fname, allow_pickle=True) as f:
        try:
            ph_arr = f['prediction_horizon_arr']
        except:
            for key in f.keys():
                if key.find('og') != -1:
                    ph_arr = f[key]
                    break
        gru_ph_arr_list.append(ph_arr)
    num_runs = ph_arr.shape[0]
    ph_arr.sort()
    gru_medians.append(ph_arr[int(0.5*num_runs-1)])
print(gru_medians)
```

```python
def gru_plot(box_or_violin):
    yticks_min = 0
    yticks_max = int(np.max(gru_ph_arr_list)+1)
    yticks_pos = np.arange(yticks_min, yticks_max+1, 1)

    # box_or_violin = 2 # 0 == box
    #                   # 1 == violin
    #                   # 2 == box with violin super-imposed
    if box_or_violin == 0:
        parts = plt.boxplot(
            gru_ph_arr_list,
        )
    elif box_or_violin == 1:
        parts = plt.violinplot(
            gru_ph_arr_list,
            showmedians=True
        )
    elif box_or_violin == 2:
        parts = plt.violinplot(
            gru_ph_arr_list,
        )
        parts['cmaxes'].set(linewidth=0.)
        parts['cmins'].set(linewidth=0.)
        parts['cbars'].set(linewidth=0.)
        # parts['medians'].set(linewidth=0.)
        parts2 = plt.boxplot(
            gru_ph_arr_list,
        )

    title_str = 'Simple RNN'

    plt.grid(True)
    plt.xticks(
        np.arange(gru_numoutsteps.shape[0]) + 1,
        ('Teacher\nforced',) + tuple([str(elem) for elem in gru_numoutsteps[1:]]),
        **xticks_kwargs
    )
    plt.yticks(yticks_pos, **yticks_kwargs)
    plt.xlabel('AR Training Steps', **xlabel_kwargs)
    plt.ylabel('Prediction Horizon [LT]', **ylabel_kwargs)
    plt.title(title_str+' Auto-regressive Training', **title_kwargs)

    save_str = title_str.replace(' ', '')
    save_str = save_str.lower()
    if box_or_violin == 0:
        savefig_name = save_str+'-study01--boxplot.pdf'
    elif box_or_violin == 1:
        savefig_name = save_str+'-study01--violinplot.pdf'
    elif box_or_violin == 2:
        savefig_name = save_str+'-study01--boxandviolinplot.pdf'

    plt.savefig(gru_plot_dir+'/'+savefig_name, dpi=300, bbox_inches='tight')
    plt.show()
    plt.close()
```

```python
for i in range(3):
    esn_plot(i)
    gru_plot(i)
```
